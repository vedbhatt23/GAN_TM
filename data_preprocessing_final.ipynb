```python
import os
import pandas as pd
from pytube import YouTube
import soundfile as sf
import librosa

def save_waveform(youtube_id, start_time, end_time, save_dir):
    """
    Extracts audio waveform from a YouTube video for a specific time range and saves it.

    Args:
        youtube_id: String containing the Youtube video ID.
        start_time: Float representing the start time in seconds.
        end_time: Float representing the end time in seconds.
        save_dir: String representing the directory to save the extracted waveform.

    Returns:
        True if successful, False otherwise.
    """
    try:
        # Create filename (use ID and timestamps for reference)
        filename = f"{youtube_id}.wav"
        save_path = os.path.join(save_dir, filename)

        # Download and extract waveform
        yt = YouTube(f"https://www.youtube.com/watch?v={youtube_id}")
        audio_stream = yt.streams.filter(only_audio=True).first().download()
        y, sr = librosa.load(audio_stream, offset=start_time, duration=end_time - start_time - 7)

        # Save waveform to .wav file
        sf.write(save_path, y, sr)  # Changed to sf.write, since sf.write_wav is deprecated

        return True
    except Exception as e:
        print(f"Error extracting waveform for {youtube_id}: {e}")
        return False

# Create directory if not found
def create_directory(directory):
    if not os.path.exists(directory):
        os.makedirs(directory)

# Assuming your DataFrame is named 'df'
save_directory = "C:/Users/CL502_04/PycharmProjects/Gans102/music_data_1" # Replace with your desired folder path

# Create directory if not found
create_directory(save_directory)

# Extract waveforms for each row
for index, row in df.iterrows():
    youtube_id = row["ytid"]
    start_time = row["start_s"]
    end_time = row["end_s"]

    success = save_waveform(youtube_id, start_time, end_time, save_directory)

    if success:
        print(f"Waveform saved for video {youtube_id} (index {index})")
    else:
        print(f"Failed to save waveform for video {youtube_id} (index {index})")
```

```python
import os
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt

# Function to convert WAV files to spectrograms and save images
def convert_to_spectrogram(wav_path, specto_folder):
    # Load the audio file
    try:
        y, sr = librosa.load(wav_path)
    except FileNotFoundError:
        print(f"File not found: {wav_path}")
        return None
    
    # Compute the spectrogram
    spectrogram = librosa.feature.melspectrogram(y=y, sr=sr)
    db_spectrogram = librosa.power_to_db(spectrogram, ref=np.max)
    
    # Save the spectrogram image with preserved aspect ratio
    specto_path = os.path.join(specto_folder, '-'+os.path.basename(wav_path).replace('.wav', '.png'))
    plt.figure(figsize=(db_spectrogram.shape[1]/100, db_spectrogram.shape[0]/100))  # Set figure size based on image shape
    plt.imshow(db_spectrogram, cmap='viridis')
    plt.gca().set_axis_off()  # Turn off axis
    plt.subplots_adjust(top=1, bottom=0, right=1, left=0, hspace=0, wspace=0)
    plt.margins(0, 0)
    plt.gca().xaxis.set_major_locator(plt.NullLocator())
    plt.gca().yaxis.set_major_locator(plt.NullLocator())
    plt.savefig(specto_path, bbox_inches='tight', pad_inches=0, format='png', dpi=100)
    plt.close()
    
    return specto_path

# Create a folder for spectrograms if it doesn't exist
specto_folder = 'specto1'
if not os.path.exists(specto_folder):
    os.makedirs(specto_folder)

# Apply the function to each WAV file and store the paths
caption_wav_df['specto_path'] = caption_wav_df['wav_path'].apply(lambda x: convert_to_spectrogram(x, specto_folder))
```

```python
from keras.layers import Input, Conv1D, LeakyReLU, Dropout, Flatten, Dense
from keras.models import Model

def build_discriminator(segment_length, num_channels=1):
    # Input for the generated or real audio segment
    audio_input = Input(shape=(segment_length, num_channels))

    # Convolutional layers
    x = Conv1D(32, kernel_size=3, strides=2, padding='same')(audio_input)
    x = LeakyReLU(alpha=0.2)(x)
    x = Dropout(0.3)(x)

    x = Conv1D(filters=64, kernel_size=3, strides=2, padding='same')(x)
    x = LeakyReLU(alpha=0.2)(x)
    x = Dropout(0.3)(x)

    # Flatten layer
    x = Flatten()(x)
    
    x = Dense(256, activation='sigmoid')(x)
    x = LeakyReLU(alpha=0.2)(x)

    # Output layer
    validity = Dense(1, activation='sigmoid')(x)

    # Discriminator model
    discriminator = Model(inputs=audio_input, outputs=validity)

    return discriminator

# Rebuild the discriminator
segment_length = 3 * 22050  # Segment length in number of samples
num_channels = 1  # Set to 2 for stereo audio
discriminator = build_discriminator(segment_length, num_channels)
discriminator.summary()

discriminator.save("discriminator_model03.h5")
```

```python
from keras.layers import Input, Conv1D, Conv2DTranspose, BatchNormalization, ReLU, Reshape, Conv2D, Concatenate, Dense, Flatten
from keras.models import Model

def build_generator(segment_length, max_caption_length, spectrogram_shape):
    # Caption input
    caption_input = Input(shape=(max_caption_length,))

    # Spectrogram input
    spectrogram_input = Input(shape=spectrogram_shape)

    # Spectrogram processing layers (Convolutional layers)
    conv1d_filters = 8
    conv_kernel_size = 3
    x_spectrogram = Conv1D(filters=conv1d_filters, kernel_size=conv_kernel_size, padding='same')(spectrogram_input)
    x_spectrogram = BatchNormalization()(x_spectrogram)
    x_spectrogram = ReLU()(x_spectrogram)

    # Flatten the spectrogram representation
    x_spectrogram_flat = Flatten()(x_spectrogram)

    # Text processing layers (Embedding layer can be added if necessary)
    x_caption = Dense(128)(caption_input)
    x_caption = ReLU()(x_caption)

    # Concatenate processed spectrogram and text representations
    x = Concatenate(axis=-1)([x_spectrogram_flat, x_caption])

    # Reshape to match Conv2DTranspose input shape
    x = Reshape((1, 1, x.shape[-1]))(x)

    # Convolutional layers
    conv2d_filters = 16
    x = Conv2DTranspose(filters=conv2d_filters, kernel_size=(1, 3), strides=(1, 1), padding='same')(x)
    x = ReLU()(x)

    x = Conv2DTranspose(filters=1, kernel_size=(1, segment_length), strides=(1, 1), padding='valid')(x)
    x = ReLU()(x)

    # Reshape to waveform
    waveform = Reshape((segment_length, 1))(x)

    # Model
    generator = Model(inputs=[caption_input, spectrogram_input], outputs=waveform)

    return generator

# Example usage
segment_length = 3 * 22050
max_caption_length = 62
spectrogram_shape = (130, 128)
generator = build_generator(segment_length, max_caption_length, spectrogram_shape)
generator.summary()

generator.save("generator_model03.h5")
```

```python
import numpy as np
import pickle
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.sequence import pad_sequences
import librosa
from keras.losses import BinaryCrossentropy
import soundfile as sf
from nltk.translate.bleu_score import corpus_bleu
import os

# Define the generator loss function
def generator_loss(y_true, y_pred):
    return BinaryCrossentropy()(y_true, y_pred)

# Load the tokenizer
with open('tokenizer10.pickle', 'rb') as handle:
    tokenizer = pickle.load(handle)

# Load the generator model
generator_model = load_model("generator_model_trained03.h5", custom_objects={'generator_loss': generator_loss})

# Tokenize and pad the captions
max_length = 62  # Match the max_caption_length used in training
sequences = tokenizer.texts_to_sequences(caption_wav_df['processed_caption'])
padded_captions = pad_sequences(sequences, maxlen=max_length, padding='post')

# Convert captions to floating-point
padded_captions = padded_captions.astype(float)

# Assuming you have the target size for the spectrogram (130, 128)
# You may adjust this according to your actual spectrogram shape
target_size = (130, 128)

# Initialize a list to store generated outputs
generated_outputs = []
reference_music_files = []

for caption, specto_path, wav_path in zip(padded_captions, caption_wav_df['specto_path'], caption_wav_df['wav_path']):
    # Load the corresponding spectrogram
    spectrogram = librosa.feature.melspectrogram(y=text.flatten(), sr=22050, n_mels=target_size[0], hop_length=512)
    spectrogram = librosa.power_to_db(spectrogram, ref=np.max)
    
    # Reshape the spectrogram to match the generator's input shape
    spectrogram = np.pad(spectrogram, ((0, 0), (0, target_size[1] - spectrogram.shape[1])), mode='constant')
    spectrogram = spectrogram.reshape(1, target_size[0], target_size[1])
    
    # Pass the caption and spectrogram through the generator
    generated_output = generator_model.predict([caption.reshape(1, max_length), spectrogram])
    
    # Reshape the generated output to remove the batch dimension
    generated_output = generated_output.reshape(-1)
    
    # Scale the generated audio back to the range of -1 to 1
    generated_output = librosa.util.normalize(generated_output)
    
    # Append the generated output to the list
    generated_outputs.append(generated_output)
    reference_music_files.append(wav_path)

# Calculate BLEU score
bleu_scores = []
for reference_music, generated_output in zip(reference_music_files, generated_outputs):
    if not os.path.exists(reference_music):
        print(f"Reference music file not found: {reference_music}. Skipping...")
        continue
    
    # Load reference music file
    reference_music, _ = librosa.load(reference_music, sr=22050)
    
    # Normalize the generated output
    generated_output_norm = librosa.util.normalize(generated_output)
    
    # Compute BLEU score
    bleu_score = corpus_bleu([[reference_music.tolist()]], [generated_output_norm.tolist()])
    bleu_scores.append(bleu_score)
    print(f"BLEU score: {bleu_score}")

# Average BLEU score
if bleu_scores:
    avg_bleu_score = np.mean(bleu_scores)
    print(f"Average BLEU score: {avg_bleu_score}")
else:
    print("No valid reference music files found.")
```
